## Answer 1
**Given:**<Br>
$ P(Fair) = 0.999$<Br>
$ P(Head | Fair) = 0.5 $<Br>

$ P(Forged) = 0.001 $<Br>
$ P(Head | Forged) = 0.9 $<Br>

**Definition:** $n$ = The number of times coin falls on "heads" out of 10 coin tosses <Br>
**According to Bayes rule:**<Br>
$ P(Forged | n=9) = \dfrac{P(n=9 | Forged) \cdot P(Forged)}{P(n=9)} $<Br><Br>
$ P(Fair | n=9) = \dfrac{P(n=9 | Fair) \cdot P(Fair)}{P(n=9)} $<Br>

**From binomial distribution:**<Br>
$ P(n=9 | Forged) = {10 \choose 9} \cdot 0.9^9 \cdot 0.1 = 0.38742$<Br>
$ P(n=9 | Fair) = {10 \choose 9} \cdot 0.5^9 \cdot 0.5 = 0.009765625$<Br>

$ P(n=9) = P(n=9 | Forged) \cdot P(Forged) +  P(n=9 | Fair) \cdot P(Fair) = 0.010143279864$<Br>

**Putting it together:**<Br>
a. **Chance this is a Forged coin:** $ P(Forged | n=9) = \dfrac{0.38742 \cdot 0.001}{0.010143279864} = 0.038195 $<Br><Br>
b. **Chance this is a Fair coin:** $ P(Fair | n=9) = \dfrac{0.009765625 \cdot 0.999}{0.010143279864} = 0.961805 $



---

```python
import math
p_fair = 0.999
p_forged = 0.001
p_n9_fair = math.comb(10, 9) * 0.5**9 * 0.5
p_n9_forged = math.comb(10, 9) * 0.9**9 * 0.1

p_n9 = p_fair * p_n9_fair + p_forged * p_n9_forged
p_forged_n9 = p_n9_forged * p_forged / p_n9
p_fair_n9 = p_n9_fair * p_fair / p_n9

p_forged_n9, p_fair_n9
```

**Using log likelihood:**<Br>
$ \theta_{ML} = \mathop{argmax} \limits_{\theta \in \Omega} LL(\theta) = $<Br>
    $ \mathop{argmax} \limits_{\theta \in \Omega} log(P(x_1,...,x_n | \theta)) = $<Br>
    $ \mathop{argmax} \limits_{\theta \in \Omega} log \prod_{i=1}^{n} (P(x_i | \theta)) = $<Br>
    $ \mathop{argmax} \limits_{\theta \in \Omega} \sum_{i=1}^{n} log (P(x_i | \theta)) = $<Br><Br>

**a. Binomial MLE:**<Br>
Suppose we got "Heads" k times in our sample. <Br>
    $ LL(\theta) = log P(x_1...x_n | \theta) = log(p^k \cdot (1-p)^{n-k}) = k \cdot log(p) + (n-k) \cdot log (1-p) $<Br>
    
Differentiate and set to Zero:<Br>
    $ \dfrac{dL(\theta)}{dp} = \dfrac{k}{p} - \dfrac{n-k}{1-p} = 0 $<Br>
    $ p = \dfrac{k}{p} $<Br><Br>

**b. Poisson MLE:**<Br>
    $ L(\theta) = L(\lambda | x_1...x_n) = \prod_{i=1}^{n} f(x_i | \lambda) = \prod_{i=1}^{n} e ^{- \lambda} \dfrac{\lambda ^{k_i}}{k_i !} $ <Br>
    $ LL(\theta) = ln L(\theta) = ln \Big( \prod_{i=1}^{n} e ^{- \lambda} \dfrac{\lambda k_i}{k_i !} \Big) = \sum_{i=1}^n ln \Big(e ^{- \lambda} \dfrac{\lambda ^{k_i}}{k_i !} \Big) = $
    $ - \lambda n -  \sum_{i=1}^n ln (x_i !) + ln \lambda \sum_{i=1}^n x_i $ <Br>

Differentiate and set to Zero:<Br>
    $ \dfrac{dL(\theta)}{d \lambda} = -m + \dfrac{1}{\lambda} \sum_{i=1}^{n} x_i = 0 $<Br>
    $ \lambda = \dfrac{1}{n} \sum_{i=1}^{n} x_i $<Br><Br>
    
**c. Normal MLE:**<Br>
    $ L(\theta) = \prod_{i=1}^n \dfrac{1}{\sqrt{2 \pi \sigma^2}} \cdot e^{-\dfrac{(x_i - \mu)^2}{2 \sigma^2}} $ <Br>
    $ LL(\theta) = -n \cdot ln(\sigma \sqrt{2 \pi}) - \sum_{i=1}^n \dfrac{(x_i - \mu)^2}{2 \sigma ^2} $<Br>
    
Differentiate and set each parameter $\mu$, $\sigma$ to Zero:<Br>
    $ \dfrac{\partial L(\theta)}{\partial \mu} = \sum_{i=1}^n \dfrac{x_i - \mu}{\sigma ^2} = 0 $ <Br>
    $ \mu = \dfrac{1}{n} \sum_{i=1}^n x_i $ <Br> <Br>
    $ \dfrac{\partial L(\theta)}{\partial \sigma} = - \dfrac{n}{\sigma} + \sum_{i=1}^n \dfrac{(x_i - \mu)^2}{2 \sigma^3} = 0 $ <Br>
    $ \sigma^2 = \dfrac{1}{n} \sum_{i=1}^n (x_i - \mu)^2 $

## Answer 5
We can define an inner product on a set of random variables using the expected value of their product: <Br>
    $ \langle X, Y \rangle := E(XY) $<Br>

We can now use the above definition with the covariance definition:<Br>
    $ Cov(X, Y) = E[(X - \mu_X) (Y - \mu_Y)] = \langle X - \mu_X, Y - \mu_y \rangle $ <Br>

Cauchy-Schwartz inequality states that:<Br>
    $ |\langle u, v \rangle|^2 \leq \langle u, u \rangle \cdot \langle v, v \rangle$<Br>
    
Then, using Cauchy-Schwartz inequality with the covariance we get: <Br>
    $ |Cov(X, Y)|^2 = $
    $ |\langle X - \mu_X, Y - \mu_Y \rangle|^2 \leq \langle X - \mu_X, X - \mu_x \rangle \langle Y - \mu_Y, Y - \mu_y \rangle = $
    $ E((X - \mu_x)^2) E((Y - \mu_Y)^2) = $
    $ Var(X) Var(Y) $<Br>

divide by the variances and take square root, we get:<Br>
    $ \dfrac{|Cov(X, Y)|}{\sqrt{Var(X)} \sqrt{Var(Y)}} \leq 1 $<Br>

    
Note that the left side of the inequality is actually the definition of pearson $\rho$ in absolute value.  so we get:<Br>
    $|\rho| \leq 1 $<Br>
    
As required.
